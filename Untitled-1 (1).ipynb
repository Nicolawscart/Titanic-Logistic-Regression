```python
data = pd.read_csv("Titanic-Dataset.csv")
```

```python
data.isnull().sum()
```

```python
data["Age"].fillna(data["Age"].median(), inplace=True)
data["Embarked"] = data.dropna(subset=['Embarked'])
data["Age"] = data["Age"].round(0)
```

```python
data.drop(["Ticket"], axis=1, inplace=True)
data.drop(["Name"], axis=1, inplace=True)
data.drop(["PassengerId"], axis=1, inplace=True)
data.drop(["Cabin"], axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
print(data["Embarked"].unique())
```

```python
data.Sex = [1 if value == "male" else 0 for value in data.Sex]
```

```python
data.Embarked = [0 if value == "C" else 1 if value == "Q" else 2 for value in data.Embarked]
```

```python
print(data["Embarked"].unique())
```

```python
#Nicolas Carter
y = data["Survived"]
x = data.drop(["Survived"], axis=1)
```

```python
#Nicolas Carter
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
#Nicolas Carter
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.30, random_state=20)
```

```python
#Nicolas Carter
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
```

```python
#Nicolas Carter
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
#Nicolas Carter
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```
